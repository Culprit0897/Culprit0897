```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Read the CSV file from local system (update the file path)
file_path = r"/content/sample_data/california_housing_test.csv"
df = pd.read_csv(file_path)
print('df',df)
```

```python
# Display basic info
print("Dataset Overview:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Basic statistics
print("\nData Statistics:")
print(df.describe())
```

```python
# Histogram of house values
plt.figure(figsize=(8, 5))
sns.histplot(df["median_house_value"], bins=50, kde=True, color="blue")
plt.xlabel("Median House Value ($)")
plt.ylabel("Frequency")
plt.title("Distribution of House Values")
plt.show()
```

```python
# Scatter plot: Median Income vs. House Value
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["median_income"], y=df["median_house_value"], alpha=0.5)
plt.xlabel("Median Income ($)")
plt.ylabel("Median House Value ($)")
plt.title("Income vs. House Value")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
